```python
import pandas as pd
import urllib
import zipfile
import warnings
warnings.filterwarnings('ignore')
```

```python
def dowonload_instruments():
    url = 'https://directlink.icicidirect.com/NewSecurityMaster/SecurityMaster.zip'
    urllib.request.urlretrieve(url, "Data/SecurityMaster.zip")        
    with zipfile.ZipFile("Data/SecurityMaster.zip", 'r') as zip_ref:
        zip_ref.extractall('Data')
    print('instruments downloaded')
```

```python
dowonload_instruments()
```

```python
def get_nearest_expiry_opt(index_name):
    global scrip
    scrip = scrip[scrip.ShortName == index_name]
    scrip = scrip[scrip.Series == 'OPTION']
    scrip.ExpiryDate = pd.to_datetime(scrip.ExpiryDate)
    scrip['ExpiryDate'] = [x.date() for x in scrip.ExpiryDate]
    return sorted(scrip.ExpiryDate)[0]

def get_nearest_expiry_fut(index_name):
    global scrip
    scrip = scrip[scrip.ShortName == index_name]
    scrip = scrip[scrip.Series == 'FUTURE']
    scrip.ExpiryDate = pd.to_datetime(scrip.ExpiryDate)
    scrip['ExpiryDate'] = [x.date() for x in scrip.ExpiryDate]
    return sorted(scrip.ExpiryDate)[0]
```
